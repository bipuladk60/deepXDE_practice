```python
import deepxde as dde
```

```python
#Step 1: Specifying the computational domain using the geometry module
geom = dde.geometry.Polygon([[0, 0], [1, 0], [1, -1], [-1, -1], [-1, 1], [0, 1]])

#Step 2: Specify the PDE using the grammar of tensorFlow
def pde(x, y):
    dy_xx = dde.grad.hessian(y, x, i=0, j=0)
    dy_yy = dde.grad.hessian(y, x, i=1, j=1)
   
    return -dy_xx - dy_yy - 1

#Step 3: Specify the boundary and initial conditions
def boundary(_, on_boundary):
    return on_boundary
bc = dde.icbc.DirichletBC(geom, lambda x: 0, boundary)


#Step 4: Combining the geometry, PDE, boundary/initial condition into data.PDE
data = dde.data.PDE(geom, pde, bc, num_domain=1200, num_boundary=120, num_test=1500)

#Step 5: Constructing the neural network
net = dde.nn.FNN([2] + [50] * 4 + [1], "tanh", "Glorot uniform")

#Step 6: Combining data.PDE and neural network into a model
model = dde.Model(data, net)

#compiling and setting optimization hyperparameters: optimizer and learning rate
model.compile("adam", lr=0.001)

#training the model using the adam optimizer
model.train(iterations=50000)

#compiling again using L-BFGS optimizer
model.compile("L-BFGS")

losshistory, train_state = model.train()
dde.saveplot(losshistory, train_state, issave=True, isplot=True)
```
